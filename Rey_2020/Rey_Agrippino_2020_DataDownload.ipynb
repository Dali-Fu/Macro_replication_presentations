# Replication data for Agrippino & Rey (2020)
#### Citation:

`Silvia Miranda-Agrippino, Hélène Rey, U.S. Monetary Policy and the Global Financial Cycle, The Review of Economic Studies, Volume 87, Issue 6, November 2020, Pages 2754–2776`

#### Data downloaded on the from the following website
http://www.helenerey.eu/RP.aspx?pid=Published-Papers_en-GB&aid=291587444_67186463733

```python
import pandas as pd
import os
```

```python
os.mkdir('Data')
```

# 1. VAR data

```python
url = 'http://www.helenerey.eu/AjaxRequestHandler.ashx?Function=GetSecuredDOC&DOCUrl=App_Data/helenerey_eu/Published-Papers_en-GB/_Documents_2019-20/291587444_67186463733_GFC_VARdata%20WEB-LITE.xlsx'
df_var = pd.read_excel(url, header=[0,1], parse_dates=[0], index_col=[0])
df_var.head()
```

```python
df_var.info()
```

```python
df_var.to_csv('Data/var_data.csv')
```

# 2. Global factor data

## 2.1 Original

```python
url = 'http://www.helenerey.eu/AjaxRequestHandler.ashx?Function=GetSecuredDOC&DOCUrl=App_Data/helenerey_eu/Published-Papers_en-GB/_Documents_2019-20/291587444_67186463733_Miranda-AgrippinoRey%20-%20Global%20Factor%20Data.xlsx'
df_factor = pd.read_excel(url, header=2)
df_factor.rename(columns={'Unnamed: 1':'Date'},inplace=True)
df_factor.head(2)
```

```python
df_factor.to_csv('Data/GF_original_data.csv')
```
